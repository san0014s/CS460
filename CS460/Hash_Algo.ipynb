```python
def hashBasedAlgo(text_file, pat, q):
    text = open(text_file, 'r', encoding = 'utf-8')
    r = text.read()
    
    m = len(pat)
    n = len(r)
    i = 0
    j = 0 
    p = 0 
    t = 0
    h = 1
    d = 256
    
    while(i<m):
        h = (h*d)%q
        i+=1

    
    #hash value for text_file and pattern
    for i in range(m):
        p = (d * p + ord(pat[i])) % q
        t = (d * t + ord(r[i])) % q
        
        
    for i in range(n-m+1):
        if(p == t):
             for j in range(m):
                    if(r[i + j] != pat[j]):
                        i += 1
                        break
        
        j += 1                      
        if(j == m):
            print("Occurence found in position", i+j)
        i += 1
            
        if(i < n-m):
            t = (d * (t - ord(r[i]) * h) + ord(r[i + m])) % q
                 
                 
            if(t<0):
                t = t + q
                 
```
